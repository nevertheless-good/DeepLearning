```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../학습모델/EfficientNet_Result.xlsx")
```

```python
df
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
from numpy import arange
epochs = range(1, 101)
 
plt.plot(epochs, df.train_loss, label='Training Loss')
plt.plot(epochs, df.valid_loss, label='Validation Loss')
 
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
 
plt.xticks(arange(0, 101, 20))
plt.yticks(arange(0, 9, 1))
 
plt.legend(loc='best')
plt.show()
```

```python
from numpy import arange
epochs = range(1, 101)
 
plt.plot(epochs, df.train_acc, label='Training Acc')
plt.plot(epochs, df.valid_acc, label='Validation Acc')
 
plt.title('Training and Validation Acc')
plt.xlabel('Epochs')
plt.ylabel('Acc')
 
plt.xticks(arange(0, 101, 20))
plt.yticks(arange(0, 101, 20))
 
plt.legend(loc='best')
plt.show()
```

```python
epochs = range(1, 31)
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(100, 50))

xlines = [10, 11, 18, 20, 30]
titles = ['train/box_loss', 'train/obj_loss', 'train/cls_loss', 'metrics/precision', 'metrics/recall',
         'val/box_loss', 'val/obj_loss', 'val/cls_loss', 'metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']

for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.title(titles[i], fontdict = {'fontsize' : 80})
    plt.plot(epochs, df_30[titles[i]], 'o-', markersize=30, linewidth=10)
    [ plt.axvline(x=l, color='r', ls='--', lw=5) for l in xlines ]
    plt.xticks(fontsize=70)
    plt.yticks(fontsize=70)
```
